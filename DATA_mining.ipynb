```python
#Calculate Lipinski descriptors

#Drug likeness is based on the Absorption, Distribution, metabolism and excretion (ADME) that is also known as the pharmacokinetic profile. Lipinki analyzed all orally active FDA-approved drugs in the formulation of what is known as the Rule-of-Five or Lipinski's Rule.

#The Lipinski's Rule stated the following:

#Molecular weight <500 Dalton

#Octanol-water partition coefficient (LogP) <5

#Hydrogen bond donors< 5

#Hydrogen bond acceptors <10
```

```python
!pip install rdkit
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
```

```python
def lipinski(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem)
        moldata.append(mol)

    baseData= np.arange(1,1)
    i=0
    for mol in moldata:

        desc_MolWt = Descriptors.MolWt(mol)
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_NumHDonors = Lipinski.NumHDonors(mol)
        desc_NumHAcceptors = Lipinski.NumHAcceptors(mol)

        row = np.array([desc_MolWt,
                        desc_MolLogP,
                        desc_NumHDonors,
                        desc_NumHAcceptors])

        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1

    columnNames = ['Molecular_Weight', 'Number of Hydrogen Donors', 'Number of Hydrogen Acceptors', 'OWPC']          #columnNames=["MW","LogP","NumHDonors","NumHAcceptors"]
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)

    return descriptors
```

```python
#To allow IC50 data to be more uniformly distributed, we will convert IC50 to the negative logarithmic scale which is essentially -log10(IC50).

#This custom function pIC50() will accept a DataFrame as input and will:

#Take the IC50 values from the standard_value column and converts it from nM to M by multiplying the value by 10

#Take the molar value and apply -log10

#Delete the standard_value column and create a new pIC50 column
```

```python
import numpy as np

def pIC50(input):
    pIC50 = []

    for i in input['standard_value_norm']:
        molar = i*(10**-9) # Converts nM to M
        pIC50.append(-np.log10(molar))

    input['pIC50'] = pIC50
    x = input.drop('standard_value_norm', 1)

    return x
```

```python
def norm_value(input):
    norm = []

    for i in input['standard_value']:
        if i > 100000000:
          i = 100000000
        norm.append(i)

    input['standard_value_norm'] = norm
    x = input.drop('standard_value', 1)

    return x
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
threshold = VarianceThreshold(threshold=.1)
```

```python
labels = shuttledata.class_label
```

```python
#Feature selection
```

```python
def variance_threshold_selector(data, threshold=0.15):
    selector = VarianceThreshold(threshold)
    selector.fit(data)
    return data[data.columns[selector.get_support(indices=True)]]
```

```python
classification_data_HV = variance_threshold_selector(shuttledata)
```

```python
classification_data_HV = pd.merge(classification_data_HV,labels,left_index=True,right_index=True)
```

```python

classification_data_HV # removing features that have a variance less than 15%
```

```python
## remove features that display high correlation with many other features
```

```python
correlated_features_2 = set()
corr_matrix_class = classification_data_HV.corr()
```

```python
for i in range(len(corr_matrix_class.columns)):
    for j in range(i):
        if abs(corr_matrix_class.iloc[i, j]) > 0.8:
            colname = corr_matrix_class.columns[i]
            correlated_features_2.add(colname)
```

```python
correlated_features_2
```

```python
#Remove correlated features.
def remove_correlated_features(features, data):
    for x in features:
        data.drop(x, axis=1, inplace=True)
    return data
```

```python
classification_data_HV = remove_correlated_features(correlated_features_2, classification_data_HV)
```

```python
# RFECV method with logistic regression estimator and 5-fold cross-validation
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
import pandas as pd

# Assuming 'X' is your feature matrix and 'y' is the target variable
# Initialize RFECV
rfecv = RFECV(estimator=LogisticRegression(), cv=5)  # Use your estimator and cv value

# Fit RFECV
rfecv.fit(X, y)

# Get the selected features using get_support() method
selected_features_mask = rfecv.get_support()

# Filter the columns based on the selected features
selected_features = X.loc[:, selected_features_mask]

# Add the target variable if needed
selected_features['class_label'] = y

# Now use these selected features for further processing or modeling
model = LogisticRegression()
model.fit(selected_features.drop('class_label', axis=1), selected_features['class_label'])
```
