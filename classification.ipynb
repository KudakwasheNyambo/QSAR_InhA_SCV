Classification modelling

```python
import pandas as pd
from numpy import where
import numpy as np
from collections import Counter
from sklearn.datasets import make_blobs
from matplotlib import pyplot
class_data = pd.read_csv('classification_ready.csv')
```

```python
class_data.head()
```

```python
X = class_data.drop(columns='class_label')
y = class_data['class_label']
print(X.shape, y.shape)
```

```python
y.value_counts()
```

SPLIT THE DATA INTO TRAINING AND TEST SETS

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,stratify=y, random_state=44) ##random_state parameter for reproducible results
```

```python
y_test.shape
X_test.shape
```

```python
X_train.to_csv('Train.csv', index=False)
```

```python
X_test.to_csv('Test1.csv', index=False)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, recall_score, precision_score, mean_squared_error, r2_score
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import roc_curve, auc
```

```python
clfmodel = make_pipeline(StandardScaler(), SVC(kernel='poly', random_state=42))
```

```python
clfmodel.fit(X_train,y_train)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
predictionsc = clfmodel.predict(X_test)
print("CLF", predictionsc)
```

```python
predictionsc = clfmodel.predict(X_train)
print("CLF", predictionsc)
```

```python
print(y_test.shape)
print(predictionsc.shape)
```

```python
predictionsc1 = clfmodel.predict(X_test)
```

```python
predictionsc1.shape
predictionsc1 = clfmodel.predict(X_test)
```

```python
uniquelabels = set(y_test)|set(predictionsc1)
```

```python
print('classificatio report:', classification_report(y_test, predictionsc1, labels=list(uniquelabels)))
```
